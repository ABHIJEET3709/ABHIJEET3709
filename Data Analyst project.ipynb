```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.read_excel("Walmart_Sales.xlsx")
```

 A. Analyze the performance of sales and revenue at the city and branch level ( 5 marks)

```python
df = pd.read_excel("Walmart_Sales.xlsx")
print(df)
```

```python
df['Total Revenue'] = df['Unit price'] * df['Quantity']

xyz = df.groupby(['City', 'Branch']).agg(
    Total_Sales=('Quantity', 'sum'),
    Total_Revenue=('Total Revenue', 'sum')
).reset_index()

print(xyz)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='Total_Revenue', hue='Branch', data=xyz)
plt.title('Total Revenue by City and Branch')
plt.xlabel('City')
plt.ylabel('Total Revenue')
plt.show()
```

B. What is the average price of an item sold at each branch of the city (10 marks)


```python
xyz['Average_Price'] = xyz['Total_Revenue'] / xyz['Total_Sales']

# Display the result
print(xyz[['City', 'Branch', 'Average_Price']])
```

C. Analyze the performance of sales and revenue, Month over Month across the
Product line, Gender, and Payment Method, and identify the focus areas to
get better sales for April 2019. (15 marks)

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

monthly_performance = df.groupby(['Year', 'Month', 'Product line', 'Gender', 'Payment']).agg(
    Total_Sales=('Quantity', 'sum'),
    Total_Revenue=('Total Revenue', 'sum')
).reset_index()

april_2019_data = monthly_performance[(monthly_performance['Year'] == 2019) & (monthly_performance['Month'] == 4)]

# Plot Revenue by Product Line
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_performance[monthly_performance['Year'] == 2019],
            x='Month', y='Total_Revenue', hue='Product line')
plt.title('Monthly Revenue by Product Line in 2019')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()

# Plot Revenue by Gender
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_performance[monthly_performance['Year'] == 2019],
            x='Month', y='Total_Revenue', hue='Gender')
plt.title('Monthly Revenue by Gender in 2019')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()

# Plot Revenue by Payment Method
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_performance[monthly_performance['Year'] == 2019],
            x='Month', y='Total_Revenue', hue='Payment')
plt.title('Monthly Revenue by Payment Method in 2019')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()
```
